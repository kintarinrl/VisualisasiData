```python
import networkx as nx
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
G.add_edges_from([("bandung", "banjar"),
                  ("bandung","bekasi"),
                  ("bandung","bogor"),
                  ("bandung","cimahi"),
                  ("bandung","cirebon"),
                  ("bandung","sukabumi"),
                  ("bandung","tasikmalaya"),
                  ("banjar","bandung"),
                  ("banjar","bekasi"),
                  ("banjar","cimahi"),
                  ("banjar","tasikmalaya"),
                  ("bekasi","bandung"),
                  ("bekasi","banjar"),
                  ("bekasi","cimahi"),
                  ("bekasi","cirebon"),
                  ("bekasi","tasikmalaya"),
                  ("bogor","depok"),
                  ("cimahi","bandung"),
                  ("cimahi","banjar"),
                  ("cimahi","bekasi"),
                  ("cimahi","cirebon"),
                  ("cimahi","tasikmalaya"),
                  ("cirebon","bandung"),
                  ("cirebon","bekasi"),
                  ("cirebon","cimahi"),
                  ("depok","bogor"),
                  ("sukabumi","bogor"),
                  ("tasikmalaya","bandung"),
                  ("tasikmalaya","banjar"),
                  ("tasikmalaya","bekasi"),
                  ("tasikmalaya","cimahi")])
nx.draw(G, with_labels = True)
plt.show()
```

```python
print("---------------\n")
print("nodes for named network:","\n",G.nodes())
print("----------------\n")
print("nodes for named network:","\n",G.edges())
```

```python
fig = plt.figure(figsize=(10,10))
nx.draw(G, with_labels=True, node_size=5000, edge_color='black')
```

```python
nx.degree_centrality(G)
```

5 kota terpenting berdasarkan nilai degree centrality: 
- Bandung
- Banjar
- Bekasi
- Bogor
- Cimahi

```python
nx.betweenness_centrality(G)
```

5 kota terpenting berdasarkan nilai betweenness centrality: 
- Bandung
- Banjar
- Bekasi
- Bogor
- Cimahi
